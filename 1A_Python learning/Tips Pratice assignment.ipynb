```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

# Imported pandas, matplot, seaborn libraries

```python
fl=pd.read_csv("tips.csv")
```

```python
fl
```

# imported file tipes and assigned as fl

```python
fl.drop(["Payer Name", "CC Number", "Payment ID"],axis=1,inplace=True)
```

Droup out unwanted column and saved the changes(by inplace=True)

```python
plt.figure(figsize=(5,5))                               #used to resize plot

sns.histplot(x="total_bill", data=fl)        #(histogram plot under lib of seaborn)
plt.xlabel("Total Bill")                     #(x y axis labeling given under lib of matplot)
plt.ylabel("Count")
plt.title("Total Bill Amt.")
plt.show()                                             
```

```python
plt.figure(figsize=(5,4))                               #used to resize plot

sns.countplot(x="smoker", data=fl)     #Show the counts of observations in smokar (count requried ony either x or y value)
```

```python
sns.violinplot(x="time",y="tip", data=fl)      #in violine plot either one axis should be numeric
```

```python
sns.stripplot(x="time",y="tip", data=fl) 
```

it is observed that most males and females are giving tips in the range of 1 to 6 dollars. only few males can be seen giving tips higher than 6 dollars and only 1 female can be observed giving a tip greater than 6 dollars.

```python
sns.swarmplot(x="time",y="tip", data=fl) 
```

from graph we can say that $2 is common tip in both dinner and lunch while tip rcieved more in dinner time

```python
sns.countplot(x="day", data=fl)
```

```python
fl["size"].unique()          # it shows uniqu observations in size column
```

```python
plt.figure(figsize=(5,5))                               #used to resize plot


sns.stripplot(x="day",y="tip",hue="sex",data=fl)       #hue used here to add 3rd category to plot as sex.
plt.xlabel="Days"
plt.ylabel="Tip$"
plt.show()
```
