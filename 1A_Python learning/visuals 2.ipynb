```python
import pandas as pd
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(["Payer Name","CC Number","Payment ID"],axis=1,inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Univariate analysis

```python
sns.histplot(x="total_bill",data=df)
plt.title("bill plot")
```

```python
sns.histplot(x="tip",data=df)
plt.title("tip plot")
```

```python
sns.countplot(x="smoker",data=df)
```

```python
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(x="day",data=df)
```

```python
df["size"].unique()
```

```python
sns.countplot(x="size",data=df)
```

Bivariate analysis

```python
sns.stripplot(x='sex',y='tip', data=df,jitter = True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

it is observed that most males and females are giving tips in the range of 1 to 6 dollars. only few males can be seen giving tips higher than 6 dollars and only 1 female can be observed giving a tip greater than 6 dollars.

Multivariate analysis

```python
plt.figure(figsize=(7,5))
sns.stripplot(x='sex', y='tip',hue='day', data=df)
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.stripplot(x='day', y='tip',hue='sex', data=df)
plt.ylabel('tip ($)')
plt.show()
```

on all days males were seen giving higher tips in comparision to females

```python
plt.figure(figsize=(7,5))
sns.stripplot(x='day', y='total_bill',hue='sex', data=df)
plt.ylabel('total bills')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=df, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```
